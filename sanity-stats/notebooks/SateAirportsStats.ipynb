# Santity Check and Statistics

```python
import os

from iquaflow.sanity import SanityCheck
from iquaflow.ds_stats import DsStats

%matplotlib inline
```

```python
data_path = os.path.join("../datasets", "SateAirportsCoco")
output_folder = "out_pru_airports"
```

## Sanity Check

- This runs a sanity check on the dataset (images and annotations)
- It also output problems list (no problems for the present case)

```python
sc = SanityCheck(data_path, output_folder)
sc.check_annotations()
```

## Statistics

- Get statistics from dataset, the output is saved in a json within the dataset
- Quick preview plots are available
- Note that "area_coverage_by_class" only works with segmentation. Thus is not available for this dataset

```python
dss = DsStats(data_path, output_folder)
stats = dss.perform_stats(show_plots = True)
```
