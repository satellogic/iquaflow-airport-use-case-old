## Load LabelMe tags jsons and create Coco style Json

```python
import json
import os
import shutil
import numpy as np

def normalize_corners(p):
    x1= round(min(p[0][0], p[1][0]))
    x2= round(max(p[0][0], p[1][0]))
    y1= round(min(p[0][1], p[1][1]))
    y2= round(max(p[0][1], p[1][1]))
    return (x1,y1), (x2,y2)

#base_path = "/Users/gbecker/Satellogic/iqf_dev/imagequalityframework/datasets/"
base_path = "../datasets/"

planes_path= base_path+"SateAirportsLabme/"
json_path= base_path+"SateAirportsLabme/"
    
id_count = 1
cat_id_count = 1
cat_dict = {}
images = []
annotations = []

for json_file in (x for x in os.listdir(json_path) if '.json' in x):  #int(x[:-5])<=84):
    f = open(json_path+json_file)
    tags = json.load(f)
    f.close()

    print(tags.keys())
    img = tags["imagePath"]
    img_id = int(img.split('.')[0])
    images.append({'file_name': img, 'height': tags["imageHeight"], 'width': tags["imageWidth"], 'id': img_id})
    
        
    for sh in tags["shapes"]: #[x for x in tags["shapes"] if x["label"]=="airplane"]:
        if not sh["label"] in cat_dict:
            cat_dict[sh["label"]] = cat_id_count
            cat_id_count += 1
        p1, p2 = normalize_corners(sh["points"])
        print(sh["label"], p1, p2)
        new_an = {'image_id': img_id,
                  'bbox': [float(p1[0]),
                           float(p1[1]),
                           float(p2[0] - p1[0]),
                           float(p2[1] - p1[1])],
                  'category_id': cat_dict[sh["label"]],
                  'id': id_count}
        annotations.append(new_an)
        id_count += 1
```

```python
print("----->  IMAGES[:20]: ", images[:20])
print("----->  ANNOTATIONS[:20]: ",annotations[0:20])
print("----->  CATEGORIES: ", cat_dict)
```

```python
categories = []
for cat in cat_dict:
    categories.append({'supercategory': cat, 'id': cat_dict[cat], 'name': cat})
print(categories)
```

```python
sate_planes_coco = {'images': images,
                    'annotations': annotations,
                    'categories': categories}
```

```python
print(sate_planes_coco['images'][0])
print(sate_planes_coco['annotations'][0])
print(sate_planes_coco['categories'][0])
```

```python
with open(base_path+'SateAirportsCoco/sate_airplanes.json', 'w') as json_file:
    json.dump(sate_planes_coco, json_file)
```
